```python
def check_df(df, tipo):
    # tipo == 'simple' - Solo muestra volúmenes y cabecera
    if tipo == 'simple':
        print("¿Cuántas filas y columnas hay en el conjunto de datos?")
        num_filas, num_columnas = df.shape
        print("\tHay {:,} filas y {:,} columnas.".format(num_filas, num_columnas))
        

        print("¿Cuáles son las primeras dos filas del conjunto de datos?")
        display(df.head(2))
        print('\n########################################################################################')

    else:
        print("¿Cuántas filas y columnas hay en el conjunto de datos?")
        num_filas, num_columnas = df.shape
        print("\tHay {:,} filas y {:,} columnas.".format(num_filas, num_columnas))
        print('\n########################################################################################')

        print("¿Cuáles son las primeras dos filas del conjunto de datos?")
        display(df.head(2))
        print('\n########################################################################################')

        print("¿Cuáles son las últimas dos filas del conjunto de datos?")
        display(df.tail(2))
        print('\n########################################################################################')

        print("¿Cómo puedes obtener una muestra aleatoria de filas del conjunto de datos?")
        display(df.sample(2))
        print('\n########################################################################################')

        print("¿Cuáles son las columnas del conjunto de datos?")
        for i in list(df.columns):
            print('\t - ' + i)
        print('\n########################################################################################')

        print("¿Cuál es el tipo de datos de cada columna?")
        print(df.dtypes)
        print('\n########################################################################################')

        print("¿Cuántas columnas hay de cada tipo de datos?")
        print(df.dtypes.value_counts())
        print('\n########################################################################################')

        print("¿Cómo podríamos obtener información más completa sobre la estructura y el contenido del DataFrame?")
        print(df.info())
        print('\n########################################################################################')

        print("¿Cuántos valores únicos tiene cada columna?")
        print(df.nunique())
        print('\n########################################################################################')

        print("¿Cuáles son las estadísticas descriptivas básicas de las columnas numéricas?")
        display(df.describe(include='all').fillna(''))
        print('\n########################################################################################')

        print("¿Hay valores nulos en el conjunto de datos?")
        print(df.isnull().sum().sort_values(ascending=False))
        print('\n########################################################################################')

        print("¿Cuál es el porcentaje de valores nulos en cada columna?")
        print(round((df.isnull().sum()/len(df)*100), 2).sort_values(ascending=False))
        print('\n########################################################################################')
```

```python
#Analisis df 2019 
print("Análisis del DataFrame:")
check_df(df, 'completo')
```

```python
# Grafico relacion puntaje sex_code y Grafico puntaje etnia 

import matplotlib.pyplot as plt
import seaborn as sns

# Distribución de puntajes de Compass por género
sns.histplot(data=df, x='DecileScore', hue='Sex_Code_Text', kde=True)
plt.title('Distribución de puntajes de Compass por género')
plt.show()

# Estadísticas descriptivas por género
stats_by_gender = df.groupby('Sex_Code_Text')['DecileScore'].describe()
print("Estadísticas por género:")
print(stats_by_gender)

# Distribución de puntajes de Compass por etnia
sns.histplot(data=df, x='DecileScore', hue='Ethnic_Code_Text', kde=True)
plt.title('Distribución de puntajes de Compass por etnia')
plt.show()

# Estadísticas descriptivas por etnia
stats_by_ethnicity = df.groupby('Ethnic_Code_Text')['DecileScore'].describe()
print("\nEstadísticas por etnia:")
print(stats_by_ethnicity)
```

## Analisis 2do DF 

```python
df_2 = pd.read_csv(r"C:\Users\frana\Desktop\^\1. Nuclio\10. Trabajo final\Datos\compas-scores-two-years.csv")
```

```python
df_2.head()
```

```python
df_2.columns
```

```python
# Distribucion de genero 
 
# Calcular la distribución de género
gender_counts = df_2['sex'].value_counts()

# Visualización con un gráfico de barras
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color='skyblue')
plt.title('Distribución de Género')
plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.xticks(rotation=0)
plt.show()

# Comparar tasas de reincidencia por género
recidivism_by_gender = pd.crosstab(df_2['sex'], df_2['is_recid'])
print(recidivism_by_gender)
```

```python
# Calcular la distribución de raza
race_counts = df_2['race'].value_counts()

# Visualización con un gráfico de barras
plt.figure(figsize=(10, 6))
race_counts.plot(kind='bar', color='lightgreen')
plt.title('Distribución de Raza')
plt.xlabel('Raza')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.show()

# Comparar tasas de reincidencia por raza
recidivism_by_race = pd.crosstab(df_2['race'], df_2['is_recid'])
print(recidivism_by_race)
```

```python
import matplotlib.pyplot as plt

# Visualización de la distribución de 'two_year_recid'
plt.figure(figsize=(8, 6))
df_2['two_year_recid'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribución de Recidiva en Dos Años')
plt.xlabel('Recidiva en Dos Años')
plt.ylabel('Cantidad de Individuos')
plt.xticks(rotation=0)
plt.show()
```

```python
recidivism_percentage = df_2['two_year_recid'].value_counts(normalize=True) * 100
print("Proporción de Recidiva en Dos Años:")
print(recidivism_percentage)
```

```python
import seaborn as sns

# Comparar la proporción de reincidencia en dos años por género
recidivism_by_gender = df_2.groupby('sex')['two_year_recid'].mean()

# Visualización con un gráfico de barras
plt.figure(figsize=(8, 6))
recidivism_by_gender.plot(kind='bar', color=['lightblue', 'salmon'])
plt.title('Proporción de Recidiva en Dos Años por Género')
plt.xlabel('Género')
plt.ylabel('Proporción de Recidiva en Dos Años')
plt.xticks(rotation=0)
plt.show()
```

```python
# Comparar la proporción de recidiva en dos años por raza
recidivism_by_race = df_2.groupby('race')['two_year_recid'].mean()

# Visualización con un gráfico de barras
plt.figure(figsize=(10, 6))
recidivism_by_race.plot(kind='bar', color='lightgreen')
plt.title('Proporción de Recidiva en Dos Años por Raza')
plt.xlabel('Raza')
plt.ylabel('Proporción de Recidiva en Dos Años')
plt.xticks(rotation=45)
plt.show()
```

## ANALISIS EN CONJUNTO DE LOS DOS DF 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Unir los DataFrames en base a Person_ID y ID para analizar en conjunto
merged_df = pd.merge(df, df_2, left_on='Person_ID', right_on='id', suffixes=('_df', '_df_2'))

# Comparar la proporción de recidiva por género
recidivism_by_gender = merged_df.groupby('Sex_Code_Text')['two_year_recid'].mean()

# Visualización con un gráfico de barras
plt.figure(figsize=(8, 6))
recidivism_by_gender.plot(kind='bar', color=['lightblue', 'salmon'])
plt.title('Proporción de Recidiva en Dos Años por Género')
plt.xlabel('Género')
plt.ylabel('Proporción de Recidiva en Dos Años')
plt.xticks(rotation=0)
plt.show()
```

```python
# Comparar la proporción de recidiva por etnia
recidivism_by_ethnicity = merged_df.groupby('Ethnic_Code_Text')['two_year_recid'].mean()

# Visualización con un gráfico de barras
plt.figure(figsize=(10, 6))
recidivism_by_ethnicity.plot(kind='bar', color='lightgreen')
plt.title('Proporción de Recidiva en Dos Años por Etnia')
plt.xlabel('Etnia')
plt.ylabel('Proporción de Recidiva en Dos Años')
plt.xticks(rotation=45)
plt.show()
```

```python
# Comparar la proporción de recidiva por estatus legal
recidivism_by_legal_status = merged_df.groupby('LegalStatus')['two_year_recid'].mean()

# Visualización con un gráfico de barras
plt.figure(figsize=(10, 6))
recidivism_by_legal_status.plot(kind='bar', color='orange')
plt.title('Proporción de Recidiva en Dos Años por Estatus Legal')
plt.xlabel('Estatus Legal')
plt.ylabel('Proporción de Recidiva en Dos Años')
plt.xticks(rotation=45)
plt.show()

```
