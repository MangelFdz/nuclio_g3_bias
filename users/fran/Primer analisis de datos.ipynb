```python
def check_df(df, tipo):
    # tipo == 'simple' - Solo muestra volÃºmenes y cabecera
    if tipo == 'simple':
        print("Â¿CuÃ¡ntas filas y columnas hay en el conjunto de datos?")
        num_filas, num_columnas = df.shape
        print("\tHay {:,} filas y {:,} columnas.".format(num_filas, num_columnas))
        

        print("Â¿CuÃ¡les son las primeras dos filas del conjunto de datos?")
        display(df.head(2))
        print('\n########################################################################################')

    else:
        print("Â¿CuÃ¡ntas filas y columnas hay en el conjunto de datos?")
        num_filas, num_columnas = df.shape
        print("\tHay {:,} filas y {:,} columnas.".format(num_filas, num_columnas))
        print('\n########################################################################################')

        print("Â¿CuÃ¡les son las primeras dos filas del conjunto de datos?")
        display(df.head(2))
        print('\n########################################################################################')

        print("Â¿CuÃ¡les son las Ãºltimas dos filas del conjunto de datos?")
        display(df.tail(2))
        print('\n########################################################################################')

        print("Â¿CÃ³mo puedes obtener una muestra aleatoria de filas del conjunto de datos?")
        display(df.sample(2))
        print('\n########################################################################################')

        print("Â¿CuÃ¡les son las columnas del conjunto de datos?")
        for i in list(df.columns):
            print('\t - ' + i)
        print('\n########################################################################################')

        print("Â¿CuÃ¡l es el tipo de datos de cada columna?")
        print(df.dtypes)
        print('\n########################################################################################')

        print("Â¿CuÃ¡ntas columnas hay de cada tipo de datos?")
        print(df.dtypes.value_counts())
        print('\n########################################################################################')

        print("Â¿CÃ³mo podrÃ­amos obtener informaciÃ³n mÃ¡s completa sobre la estructura y el contenido del DataFrame?")
        print(df.info())
        print('\n########################################################################################')

        print("Â¿CuÃ¡ntos valores Ãºnicos tiene cada columna?")
        print(df.nunique())
        print('\n########################################################################################')

        print("Â¿CuÃ¡les son las estadÃ­sticas descriptivas bÃ¡sicas de las columnas numÃ©ricas?")
        display(df.describe(include='all').fillna(''))
        print('\n########################################################################################')

        print("Â¿Hay valores nulos en el conjunto de datos?")
        print(df.isnull().sum().sort_values(ascending=False))
        print('\n########################################################################################')

        print("Â¿CuÃ¡l es el porcentaje de valores nulos en cada columna?")
        print(round((df.isnull().sum()/len(df)*100), 2).sort_values(ascending=False))
        print('\n########################################################################################')
```

```python
#Analisis df 2019 
print("AnÃ¡lisis del DataFrame:")
check_df(df, 'completo')
```

## Analisis 2do DF 

```python
df_2 = pd.read_csv(r"C:\Users\frana\Desktop\^\1. Nuclio\10. Trabajo final\Datos\compas-scores-two-years.csv")
```

```python
# Distribucion de genero 
 
# Calcular la distribuciÃ³n de gÃ©nero
gender_counts = df_2['sex'].value_counts()

# VisualizaciÃ³n con un grÃ¡fico de barras
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color='skyblue')
plt.title('DistribuciÃ³n de GÃ©nero')
plt.xlabel('GÃ©nero')
plt.ylabel('Cantidad')
plt.xticks(rotation=0)
plt.show()

# Comparar tasas de reincidencia por gÃ©nero
recidivism_by_gender = pd.crosstab(df_2['sex'], df_2['is_recid'])
print(recidivism_by_gender)
```

```python
# Calcular la distribuciÃ³n de raza
race_counts = df_2['race'].value_counts()

# VisualizaciÃ³n con un grÃ¡fico de barras
plt.figure(figsize=(10, 6))
race_counts.plot(kind='bar', color='lightgreen')
plt.title('DistribuciÃ³n de Raza')
plt.xlabel('Raza')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.show()

# Comparar tasas de reincidencia por raza
recidivism_by_race = pd.crosstab(df_2['race'], df_2['is_recid'])
print(recidivism_by_race)
```

```python
import matplotlib.pyplot as plt

# VisualizaciÃ³n de la distribuciÃ³n de 'two_year_recid'
plt.figure(figsize=(8, 6))
df_2['two_year_recid'].value_counts().plot(kind='bar', color='skyblue')
plt.title('DistribuciÃ³n de Recidiva en Dos AÃ±os')
plt.xlabel('Recidiva en Dos AÃ±os')
plt.ylabel('Cantidad de Individuos')
plt.xticks(rotation=0)
plt.show()
```

```python
recidivism_percentage = df_2['two_year_recid'].value_counts(normalize=True) * 100
print("ProporciÃ³n de Recidiva en Dos AÃ±os:")
print(recidivism_percentage)
```

```python
import seaborn as sns

# Comparar la proporciÃ³n de reincidencia en dos aÃ±os por gÃ©nero
recidivism_by_gender = df_2.groupby('sex')['two_year_recid'].mean()

# VisualizaciÃ³n con un grÃ¡fico de barras
plt.figure(figsize=(8, 6))
recidivism_by_gender.plot(kind='bar', color=['lightblue', 'salmon'])
plt.title('ProporciÃ³n de Recidiva en Dos AÃ±os por GÃ©nero')
plt.xlabel('GÃ©nero')
plt.ylabel('ProporciÃ³n de Recidiva en Dos AÃ±os')
plt.xticks(rotation=0)
plt.show()
```

```python
# Comparar la proporciÃ³n de recidiva en dos aÃ±os por raza
recidivism_by_race = df_2.groupby('race')['two_year_recid'].mean()

# VisualizaciÃ³n con un grÃ¡fico de barras
plt.figure(figsize=(10, 6))
recidivism_by_race.plot(kind='bar', color='lightgreen')
plt.title('ProporciÃ³n de Recidiva en Dos AÃ±os por Raza')
plt.xlabel('Raza')
plt.ylabel('ProporciÃ³n de Recidiva en Dos AÃ±os')
plt.xticks(rotation=45)
plt.show()
```

## ANALISIS EN CONJUNTO DE LOS DOS DF 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Unir los DataFrames en base a Person_ID y ID para analizar en conjunto
merged_df = pd.merge(df, df_2, left_on='Person_ID', right_on='id', suffixes=('_df', '_df_2'))

# Comparar la proporciÃ³n de recidiva por gÃ©nero
recidivism_by_gender = merged_df.groupby('Sex_Code_Text')['two_year_recid'].mean()

# VisualizaciÃ³n con un grÃ¡fico de barras
plt.figure(figsize=(8, 6))
recidivism_by_gender.plot(kind='bar', color=['lightblue', 'salmon'])
plt.title('ProporciÃ³n de Recidiva en Dos AÃ±os por GÃ©nero')
plt.xlabel('GÃ©nero')
plt.ylabel('ProporciÃ³n de Recidiva en Dos AÃ±os')
plt.xticks(rotation=0)
plt.show()
```

```python
# Comparar la proporciÃ³n de recidiva por etnia
recidivism_by_ethnicity = merged_df.groupby('Ethnic_Code_Text')['two_year_recid'].mean()

# VisualizaciÃ³n con un grÃ¡fico de barras
plt.figure(figsize=(10, 6))
recidivism_by_ethnicity.plot(kind='bar', color='lightgreen')
plt.title('ProporciÃ³n de Recidiva en Dos AÃ±os por Etnia')
plt.xlabel('Etnia')
plt.ylabel('ProporciÃ³n de Recidiva en Dos AÃ±os')
plt.xticks(rotation=45)
plt.show()
```

```python
# Comparar la proporciÃ³n de recidiva por estatus legal
recidivism_by_legal_status = merged_df.groupby('LegalStatus')['two_year_recid'].mean()

# VisualizaciÃ³n con un grÃ¡fico de barras
plt.figure(figsize=(10, 6))
recidivism_by_legal_status.plot(kind='bar', color='orange')
plt.title('ProporciÃ³n de Recidiva en Dos AÃ±os por Estatus Legal')
plt.xlabel('Estatus Legal')
plt.ylabel('ProporciÃ³n de Recidiva en Dos AÃ±os')
plt.xticks(rotation=45)
plt.show()

```
